### <center> 1. Werken met input van de gebruiker

Input vragen van de gebruiker maakt je programma _interactief_ . In Python kan dit via de functie _input()_

````python
input()
````

````python
prompt = "Geef je input hier: "
input(prompt)
````

````python
prompt = "Geef je input hier: "
ingelezen = input(prompt)
print("\nJij zei net :\n" + ingelezen) # \n newline en + concatenatie
````

**Voorbeeld:** Vraag de gebruiker zijn naam en print deze terug uit in hoofdletters.

````python
prompt = "Wat is je naam? "
ingelezen = input(prompt)
print("Je naam in hoofdletters:\n" + ingelezen.upper())
````

**Voorbeeld:** Vraag de gebruiker zijn leeftijd en bereken hieruit zijn geboortejaar.

````python
import datetime
prompt = "Wat is je leeftijd? "
ingelezen = input(prompt)
huidig_jaar = datetime.date.today().year 
print(huidig_jaar)
print("Je bent geboren in het jaar:\n" + (huidig_jaar - ingelezen))
````

### <center> 2. Converteren van strings

Wanneer je input inleest van de gebruiker zal de functie _input()_ steeds een string teruggeven als resultaat, ook als de gebruiker een getal inleest. 

De variabele _ingelezen_ van vorige oef bevat dus de string  &nbsp; <kbd> "25" </kbd> &nbsp; 
en niet het getal &nbsp; <kbd> 25 </kbd> &nbsp;.

We moeten de string dus eerst omzetten naar een getal vooraleer we er numeriek mee kunnen rekenen. Dit doen we via de functies _int()_ en _float()_

Doen we dit niet, dan kan dit verassende resultaten geven:

````python
ingelezen + ingelezen
````

````python
ingelezen * 2
````

````python
ingelezen * ingelezen
````

### <center> Van strings naar getallen

- _int()_ zal een string omzetten naar een integer getal
- _float()_ zal een string omzetten naar een reel getal

Deze omzettingen gebeuren alleen als de string ook effectief een getal voorstelt

````python
int("35")
````

````python
huidig_jaar - int(ingelezen)
````

````python
print(int('11'))
print(int('11', base=10))  # 10 is already default
print(int('11', base=2))  # Binary
print(int('11', base=8))  # Octal
print(int('11', base=16))  # Hex
````

````python
number = "nonsens"
int(number)
````

````python
prompt = "Geef een komma-getal: "
getal = input(prompt)
print(type(getal))
print(float(getal))
print(type(float(getal)))
````

### <center> Van getallen naar strings
    
Omgekeerd ga je merken dat getallen soms ook omgezet moeten worden naar een string, wil je dit getal opnemen op een bepaalde plaats in een stuk tekst:

````python
import datetime
prompt = "Wat is je leeftijd? "
ingelezen = input(prompt)
huidig_jaar = datetime.date.today().year 
geboortejaar = huidig_jaar - int(ingelezen)
print("Je bent geboren in het jaar:\n" + geboortejaar)
````

````python
prompt1 = "Wat is je naam?"
ingelezen1 = input(prompt1)
prompt2 = "Wat is je leeftijd? "
ingelezen2 = input(prompt2)
geboortejaar = huidig_jaar - int(ingelezen2)
print("Dag " + ingelezen1 + ". Je bent geboren in het jaar: " + 
      str(geboortejaar) + ".")
````

### <center> 3. Formateren van strings

De zin :

```
"Dag " + ingelezen1 + ". Je bent geboren in het jaar: " + str(geboortejaar) + "."

```

noemt men <span style="color:green;"> string interpolation </span>. De variabelen worden op een bepaalde plaats in de string geplaatst. 

In dit voorbeeld via de _+_ operator, de string concatenatie, maar er zijn andere eenvoudigere manieren om dit te bekomen via <span style="color:green;"> f-strings </span>, ook formatted string literals.


````python
f'Dag {ingelezen1}. Je bent geboren in het jaar: {geboortejaar}.'
````

---
- Een f-string start dus met de letter f gevolgd door een enkele of dubbele quote

- de variabelen die je gebruikt in een f-string hoeven niet vooraf omgezet worden naar een string, je moet ze wel tussen accolades {} plaatsen.

---

````python
f'Je kan er zelfs bewerkingen zoals (2024 - 69) = \
{huidig_jaar - int(ingelezen2)} in opnemen. \
Alhoewel het niet aangeraden wordt om complexe bewerkingen \
in f-strings op te nemen.'
````
